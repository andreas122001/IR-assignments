## Now use the computation method to calculate the time period of dissimilarities between two terms

### French presidents:


Tip: use start year of 2000 and end year of 2019

```python
def KL(df, term1, term2):
  return pd.DataFrame(np.log(df[term1] / df[term2]) * df[term1])
```

```python
# Have your code here
from google_pygram import GooglePyGram as gpg
terms = ["French presidents *"]

pygram = gpg(
    corpus='English',
    corpus_year=2019,
    start_year=2000,
    end_year=2019,
    smoothing=3,
    case_sensitive=False,
    phrases=terms
)

df = pygram.to_df()
df = df.drop(columns=terms) # Remove the term with wildcard
df.plot(x="year") # Plot frequencies
```

```python
term1 = "French presidents and"
term2 = "French presidents have"

# Calculate KI Divergence
dissimilarity = KL(df, term1, term2)
np_arr = dissimilarity.to_numpy()

# Find where KI Divergence crosses zero (on the right side)
zero_crossings = np.where([np_arr[i-1]*np_arr[i] < 0 for i in range(1,len(np_arr))])[0] + 1

# To make it easy to visualize, we just calculate each year where a term becomes more popular
print(f"Term {'1' if np_arr[0] > 0 else '2'} is more popular in {dissimilarity.index[0]}")
for i in zero_crossings:
  year, val = dissimilarity.index[i], dissimilarity.iloc[i][0]
  print(f"Term {'1' if val > 0 else '2'} became more popular in {year}")

# And finally, plot the dissimilarity for better visualization
dissimilarity.plot()
```
